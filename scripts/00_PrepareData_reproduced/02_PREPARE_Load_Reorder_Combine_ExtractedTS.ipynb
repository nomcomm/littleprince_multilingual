#### Set the main path to the data and find the names of the files in order to get the subject names

```python
path              = "../../data_lpp/fmri_data/extract_ts/lpp_cn_fmri"

search_path = path + "/**/sub*" + "_LTR_Z_THP001_LP012_shen268_pauli16_aan9.npy"
print(search_path)
text_files = glob.glob(search_path, recursive = True)
#print(text_files[:2])
print(len(text_files))
text_files.sort()

subjs = []
for i in range(len(text_files)):
        subjs.append(text_files[i][45:54])
subjs = list(set(subjs))
print(subjs)
print(len(subjs))     
n_subjs = len(subjs)
```

#### Extract subject runs
Note: this is a bit cumbersome. The issue is that the run-names have varying numbers for different subjects, such as sub001_run06 being the first run (corresponding to the first sound file from LPP), whereas sub002_run09 could be the first one for that subject (this already originated at the data acquisition stage, such as taking multiple shimming scans or such). In any case, we will build a dictionary that holds the corresponding information, so that we can simply get the run-number for the first story-listening run, and build the corresponding filename as we need it.

```python
subj_run_dict ={}

for curr_sub in range(n_subjs):
    #print(curr_sub)
    search_path = path + "/**/" + subjs[curr_sub] + "*_LTR_Z_THP001_LP012_shen268_pauli16_aan9.npy"
    #print(search_path)
    text_files = glob.glob(search_path, recursive = True)
    #print(text_files[:2])
    #print(len(text_files))
    text_files.sort()

    subjs_runs = []

    for i in range(len(text_files)):
        subjs_runs.append(text_files[i][70:72])
        
    subj_run_dict[subjs[curr_sub]] = subjs_runs
```

```python
import matplotlib.pyplot as plt
import scipy

plt.figure(figsize = (20,2))
plt.plot(scipy.stats.zscore(fr1), color = 'blue')
plt.plot(scipy.stats.zscore(fr2), color = 'lightblue')

plt.plot(scipy.stats.zscore(cn1) + 6, color = 'red')
plt.plot(scipy.stats.zscore(cn2) + 6, color = 'salmon')

plt.plot(scipy.stats.zscore(en1) + 12, color = 'green')
plt.plot(scipy.stats.zscore(en2) + 12, color = 'lightgreen')

plt.ylim([-4,16])
```

```python
all = np.stack([  scipy.stats.zscore(fr1),
            scipy.stats.zscore(fr2),
            scipy.stats.zscore(cn1),
            scipy.stats.zscore(cn2),
            scipy.stats.zscore(en1),
            scipy.stats.zscore(en2)] )
corrmat =  np.corrcoef(all)



corrmat[np.triu_indices(corrmat.shape[0], 0)] = 0


plt.imshow(corrmat, cmap = 'seismic', vmin = 0.02, vmax = 0.2)
plt.colorbar()
```
